```python
import pynq.iop
from time import sleep
from pynq import Overlay
from pynq import MMIO
from pynq.board import LED
from pynq.board import RGBLED
from pynq.board import Button
from pynq.board import Switch
from pynq.iop import request_iop
from pynq.iop import iop_const
from pynq.iop import Pmod_IO
from pynq.iop import grove_th02
from pynq.iop import PMODA
from pynq.iop import PMODB
from pynq.iop import PMOD_GROVE_G3
from pynq.iop import ARDUINO
from pynq.iop import ARDUINO_GROVE_I2C
import asyncio
from IPython.display import display
from ipywidgets import widgets

ol3 = Overlay("base.bit")
ol3.download()

th02 = pynq.iop.grove_th02.Grove_TH02(PMODA, PMOD_GROVE_G3)
#nnnnnn = grove_th02(PMODA, PMOD_GROVE_G3)

MAX_BUTTONS = 4
MAX_SWITCHES = 2

switch = [0] * MAX_SWITCHES
buttons = [0] * MAX_BUTTONS
rgbled = [0] * 2

rgbled[0] = RGBLED(4)
rgbled[1] = RGBLED(5)
rgbled[0].off()
rgbled[1].off()

for i in range(MAX_BUTTONS):
    buttons[i] = Button(i)
for i in range(MAX_SWITCHES):
    switch[i] = Switch(i)
alarmButton = widgets.Button(
                description = 'Toggle Standby',
                disabled = False,
                button_style = '', #'success', 'info', 'warning', 'danger' or ''
                tooltip = 'Click me!',
                icon = 'check'
)
display(alarmButton)

swStandby = False
def toggleStandby(b):
        if swStandby:
            print("swStandby = False")
            swStandby = False
        else:
            print("swStandby = True")
            swStandby = True
        return
alarmButton.on_click(toggleStandby)
print("Starting detection...")

while True:
    sleep(0.5)
    if (switch[0].read() == 0 and switch[1].read() == 0) or swStandby:
        #initiate standby
        sleep(0.2)
        continue
    if switch[0].read() == 1 and switch[1].read() == 0:
        #monitoring on, alarm off
        '''while True:
            if th02.read()==1:
                #green light on
                rgbled[0].on(2)
                rgbled[1].on(2)
                sleep(0.1)
            else:
                rgbled[0].off()
                rgbled[1].off()
                break
        '''
        continue
    if switch[0].read() == 0 and switch[1].read() == 1:
        #standby
        sleep(0.2)
        continue
    if switch[0].read() == 1 and switch[1].read() == 1:
        #alarm on
        tuple = th02.read()
        tuplealarm = (30, 100)
        tuplestop = (0, 0)
        print("The (Temperature, Humidity) values are:", tuple)
        
        #Alarm for (Temperature, Humidity) values >= (30, 100).
        while th02.read() >= tuplealarm:
            rgbled[0].on(4)
            rgbled[1].on(4)
            sleep(0.05)
            rgbled[0].off()
            rgbled[1].off()
            sleep(0.05)
            
        #Failsafe to end the process by removing the sensor.
            if th02.read() <= tuplestop:
                break
        
    
#led.off()
rgbled[0].off()
rgbled[1].off()
del th02
#del led
del rgbled
del buttons
del ol3
```
